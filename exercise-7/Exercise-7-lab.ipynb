```python
# read in the file and make dictionaries of DBH at each census

in_file = open("/home/eeb177-student/Desktop/eeb-177/class-assignments/classwork-Tuesday-2-21/cocoli.txt")
sizes_94 = {}
sizes_97 = {}
sizes_98 = {}

#get rid of header row
in_file.readline()

for line in in_file:
    items =line.split()
    # retrrieve relevant columns
    tree_id = items[0]
    s94 = float(items[4])
    s97 = float(items[5])
    s98 = float(items[6])
    
    #populate dictionary
    sizes_94[tree_id] = s94
    sizes_97[tree_id] = s97
    sizes_98[tree_id] = s98
    
    #need to convert 'strings' to ints

in_file.close()
```

```python
import matplotlib.pyplot as plt

#let's make a histogram of DBHs in 1994
#matplotlib works on lists, so lets make one

sizes_in_1994 = list(sizes_94.values())

# sizes_in_1994
```

```python
plt.hist(sizes_in_1994, log = True)
plt.show()

sizes_in_1994 = list(sizes_94.values())#create a log transformation which helps visualize data via exponents
```

```python
sizes_in_1994 = list(sizes_94.values())
sizes_in_1998 = list(sizes_98.values())

f, (ax1, ax2) = plt.subplots(1,2,sharex = True, sharey = True)
ax1.hist(sizes_in_1994, log = True)
ax1.set_title('Hist in 1994')
ax2.hist(sizes_in_1998, log = True)
ax2.set_title('Hist in 1998')
plt.show()
```

```python
sizes_in_1994[:] = (value for value in sizes_in_1994 if value != -2)
#list comprehension

plt.hist(sizes_in_1994, log = False)
plt.show()

#list editor, to take out -2
```
